```python
import os
import pandas as pd
import numpy as np
os.chdir('/content/gdrive/My Drive/kaggle/IEEE/test')
df=pd.read_csv("test_transaction.csv",engine='python',chunksize=10000)
df1=pd.read_csv("test_identity.csv",engine='python')
training_data=[]
from tqdm import tqdm
for idf in tqdm(df):
  mean_transcation=np.array(idf.fillna(0).iloc[:,16:].mean(axis=1))
  std_transcation=np.array(idf.fillna(0).iloc[:,16:].std(axis=1))
  max_transcation=np.array(idf.fillna(0).iloc[:,16:].max(axis=1))
  min_transcation=np.array(idf.fillna(0).iloc[:,16:].min(axis=1))
  X=pd.DataFrame(idf.iloc[:,:17])
  X["mean_values"]=mean_transcation
  X["std_values"]=mean_transcation
  X["max_values"]=max_transcation
  X["min_values"]=min_transcation
  training_data.append(np.array(X))
training_data=np.array(training_data)
final_trans=[]
for i in training_data:
  for a in i:
    final_trans.append(a)
    

transcation_ids=list(idf.columns.tolist()[:17])
transcation_ids.append("mean_transcation")
transcation_ids.append("std_transcation")
transcation_ids.append("max_transcation")
transcation_ids.append("min_transcation")


df_new=pd.DataFrame(final_trans,columns=transcation_ids)
final_df=df_new.merge(df1, left_on="TransactionID", right_on="TransactionID",how="outer")
final_df = final_df.fillna(final_df.mode().iloc[0])

```

```python
os.chdir('/content/gdrive/My Drive/kaggle/IEEE/test')
final_df.to_csv("final_df_test.csv")
```

```python

import os
import pandas as pd
import numpy as np
os.chdir('/content/gdrive/My Drive/kaggle/IEEE/test')
final_df=pd.read_csv("final_df_test.csv",index_col="TransactionID")
```

```python
from sklearn import preprocessing
import pickle
os.chdir('/content/gdrive/My Drive/kaggle/IEEE/train')
a=-1
for i in final_df.dtypes.index:
  a+=1
  if final_df.dtypes.iloc[a]=="object":
    try:
      output = open('{} encoder.pkl'.format(i), 'rb')
      le = pickle.load(output)
      output.close()
      final_df[i] = le.transform(final_df[i])
    except:
      le = preprocessing.LabelEncoder()
      final_df[i] = le.fit_transform(final_df[i])
```
